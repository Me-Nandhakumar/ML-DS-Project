```python
import sklearn
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l=LabelEncoder()
a["Gender"]=l.fit_transform(a["Gender"])
a["Gender"]
a["Result"]=l.fit_transform(a["Result"])
a["Result"]
```

```python
a.tail(50)
```

```python
x=a[["Student_ID","Age","Gender","Attendance_percentage","Study_Hours_per_day","Backlogs","Sleep_Hours"]]
y=a["Result"]
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
ss_train=s.fit_transform(x_train)
ss_train
```

```python
xx_test=s.transform(x_test)
xx_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier(n_neighbors=5,metric="euclidean")
kn.fit(ss_train,y_train)
```

```python
y_pred=kn.predict(xx_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_pred,y_test)
```
